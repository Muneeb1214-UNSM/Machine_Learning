```python
for i in range(1, 55431):
  # Correctly format the URL to include the page number
  response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=b2cfe54278fd8a3d7a1266cc5138aeea&page={}'.format(i))
  # Check for successful response and 'results' key
  if response.status_code == 200 and 'results' in response.json():
    temp_df = pd.DataFrame(response.json()['results'])
    # Use pd.concat instead of the deprecated df.append
    df = pd.concat([df, temp_df], ignore_index=True)
  else:
    print(f"Could not retrieve data for page {i}. Status code: {response.status_code}")
    break # Stop if there's an issue or no more results
```
